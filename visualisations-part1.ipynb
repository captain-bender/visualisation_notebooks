```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
passengers_2019 = [5927543, 5482288, 6527489, 6798212, 6768903, 7246563, 7754139, 7680327, 6777115, 6994212, 6233801, 6696079]
passengers_2020 = [6099370, 5442081, 3106845, 206600, 228018, 350283, 866655, 1418683, 1256479, 1245624, 746954, 1143417]
plt.plot(passengers_2019)
plt.show()
```

```python
plt.plot(months, passengers_2019)
plt.show()
```

```python
plt.plot(months, passengers_2019, "o-r") #o:indicate the marker, -:indicate the line and the r:indicate the colour
plt.show()
```

```python
plt.plot(months, passengers_2019, "o-b")
plt.title("Passengers' traffic in Heathrow 2019")
plt.xlabel("Month")
plt.ylabel("Passengers")
plt.show()
```

```python
plt.plot(months, passengers_2019, "o:b")
plt.plot(months, passengers_2020, "x--r")
plt.title("Passengers' traffic in Heathrow")
plt.xlabel("Month")
plt.ylabel("Passengers")
plt.grid(True)
plt.legend(["2019", "2020"])
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(months, passengers_2019, "o:b")
ax.plot(months, passengers_2020, "x--r")
ax.set_title("Passengers' traffic in Heathrow")
ax.set_xlabel("Month")
ax.set_ylabel("Passengers")
ax.grid(True)
ax.legend(["2019", "2020"])

plt.show()
```

```python
fig, axs = plt.subplots(1, 2)

axs[0].plot(months, passengers_2019, "o:b")
axs[1].plot(months, passengers_2020, "x--r")
fig.suptitle("Passengers' traffic in Heathrow")
axs[0].set_xlabel("Month")
axs[1].set_xlabel("Month")
axs[0].set_ylabel("Passengers")
axs[1].set_ylabel("Passengers")
axs[0].grid(True)
axs[1].grid(True)
axs[0].set_title("2019")
axs[1].set_title("2020")

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

fig.suptitle("Passengers' traffic in Heathrow")

axs[0].plot(months, passengers_2019, "o:b")
axs[0].set_xlabel("Month")
axs[0].set_ylabel("Passengers")
axs[0].grid(True)
axs[0].set_title("2019")

axs[1].plot(months, passengers_2020, "x--r")
axs[1].set_xlabel("Month")
axs[1].set_ylabel("Passengers")
axs[1].grid(True)
axs[1].set_title("2020")

plt.show()
```

```python
modules=["AAC","AD","AIMT","CCA","DAOM","DASC","DCAS"]
class_students=[45, 25, 35, 40, 30, 56, 98]

plt.pie(class_students, labels=modules, autopct="%0.1f%%", explode=(0.1, 0, 0, 0, 0, 0, 0))
plt.title("Students' registration in ADTM modules")
plt.show()
```

```python
modules=["AAC","AD","AIMT","CCA","DAOM","DASC","DCAS"]
class_students=[45, 25, 35, 40, 30, 56, 98]

# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05)

plt.pie(class_students, labels=modules, autopct="%0.1f%%", explode=explode, pctdistance=0.8)

# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='w')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
plt.title("Students' registration in ADTM modules")
plt.show()
```
