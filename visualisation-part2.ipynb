```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data for each class
class_A_scores = [75, 77, 80, 90, 95]
class_B_scores = [70, 75, 80, 85, 90]
class_C_scores = [65, 70, 80, 85, 90]

# Create a boxplot
plt.boxplot([class_A_scores, class_B_scores, class_C_scores], labels=['Class A', 'Class B', 'Class C'])

# Add title and labels
plt.title('Exam Scores by Class')
plt.xlabel('Classes')
plt.ylabel('Scores')

# Show plot
plt.show()
```

```python
# Data
study_time = np.array([2, 3, 5, 7, 8])
exam_scores = np.array([70, 75, 85, 90, 95])

# Calculate regression line
m, b = np.polyfit(study_time, exam_scores, 1)

# Plot scatter plot
plt.scatter(study_time, exam_scores, color='blue', label='Data Points')

# Plot regression line
plt.plot(study_time, m * study_time + b, color='red', label=f'Regression Line: y={m:.2f}x+{b:.2f}')

# Add labels and legend
plt.xlabel('Study Time (hours)')
plt.ylabel('Exam Score')
plt.title('Study Time vs. Exam Score')
plt.legend()

# Show plot
plt.show()
```

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Add color bar
plt.colorbar(im, ax=ax)

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
